```python
# merging from 2 data fram products and order
df_merged_prods_order_inner = df_prod_clean.merge(df_ords_combined, on = 'product_id', how = 'inner',indicator='exists')
```

```python
#counting the merge flag with exist. since we can't use indicators = true because it's already included in the data we have to use indicator exist. 
df_merged_prods_order_inner['exists'].value_counts()
```

```python
# to confirm the results of the merge flag using outer
df_merged_prods_order_outer = df_prod_clean.merge(df_ords_combined, on = 'product_id', how = 'outer',indicator='exists')
```
