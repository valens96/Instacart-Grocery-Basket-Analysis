# I feel that doing it on the entire dataframe is messy where there are too many data which could prone to mistake. Whereas making it subsets are easier to read

```python
# Creating another column in the table with the maximum order by a specific user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

# Question 6

```python
# Creating another column in the table base on the specific price
ords_prods_merge['max_spender'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Categorise customer as low spender if their mean less than 10
ords_prods_merge.loc[ords_prods_merge['max_spender'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
# Categorise customer as high spender if their mean more than 10
ords_prods_merge.loc[ords_prods_merge['max_spender'] >= 10, 'spending_flag'] = 'High Spender'
```

# Question 7

```python
# Create a frequency flag base on the order behavior and find the mdeian in the days_since_prior_order"
ords_prods_merge['median_of_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
ords_prods_merge.loc[ords_prods_merge['median_of_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.
ords_prods_merge.loc[((ords_prods_merge['median_of_days_since_prior_order'] > 10) & (ords_prods_merge['median_of_days_since_prior_order'] <= 20)) , 'frequency_flag'] = 'Regular Customer'
```

```python
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.
ords_prods_merge.loc[ords_prods_merge['median_of_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
